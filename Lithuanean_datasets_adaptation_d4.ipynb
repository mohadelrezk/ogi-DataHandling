```python
#pulling csv data from file
import csv
filename='Source_datasets/Exported matrix table_d4.csv'
with open(filename, 'r') as f:
    data = list(csv.reader(f))
#print (data)
#print (f)
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#remove aggregated rows, as aggregation will take place later e.g. in pivotjs
keysList=['Republic','county']
tobedeletedList=[]
rangeCurser={}#to tag _ layers 
updated_lengh=len(data)
for row_num in range(0,updated_lengh):
    for key in keysList:
        #print (key)
        if key in data[row_num][1]:
            tobedeletedList.append(row_num)
            if 'Republic' in key:
                rangeCurser[data[row_num][0]]=row_num
                
import operator

sorted_rangeCurser = sorted(rangeCurser.items(), key=operator.itemgetter(1))

print (tobedeletedList)
print (rangeCurser)
print('sorted cursor:{}'.format(sorted_rangeCurser))


for aggregated_row_num in tobedeletedList:
    #print (data[aggregated_row_num])
    data[aggregated_row_num]=['']
    #tobedeletedList[:] = [x - 1 for x in tobedeletedList]
   #print (tobedeletedList)
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#remove first layer as it is aggregated (total by employee)
totalbyEmployeeLayer_start_end=[sorted_rangeCurser[0][1],sorted_rangeCurser[1][1]]
print (totalbyEmployeeLayer_start_end)

del_curser=totalbyEmployeeLayer_start_end[1]-1 #-1 is important not to delete the start of the next layer
print (del_curser)

for i in range(totalbyEmployeeLayer_start_end[0],totalbyEmployeeLayer_start_end[1]):
    print ("delete row {}:{}".format(del_curser,data[del_curser]))
    data[del_curser]=['']
    del_curser=del_curser-1
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#add missing ranges layers values
first_Layer_curser=[sorted_rangeCurser[1][1],sorted_rangeCurser[2][1]]
first_Layer_tag=sorted_rangeCurser[1][0]

second_Layer_curser=[sorted_rangeCurser[2][1],sorted_rangeCurser[3][1]]
second_Layer_tag=sorted_rangeCurser[2][0]

third_Layer_curser=[sorted_rangeCurser[3][1],sorted_rangeCurser[4][1]]
third_Layer_tag=sorted_rangeCurser[3][0]

fourth_Layer_curser=[sorted_rangeCurser[4][1],sorted_rangeCurser[5][1]]
fourth_Layer_tag=sorted_rangeCurser[4][0]

fifth_Layer_curser=[sorted_rangeCurser[5][1],sorted_rangeCurser[6][1]]
fifth_Layer_tag=sorted_rangeCurser[5][0]

sixth_Layer_curser=[sorted_rangeCurser[6][1],sorted_rangeCurser[7][1]]
sixth_Layer_tag=sorted_rangeCurser[6][0]

seventh_Layer_curser=[sorted_rangeCurser[7][1],sorted_rangeCurser[8][1]]
seventh_Layer_tag=sorted_rangeCurser[7][0]

eight_Layer_curser=[sorted_rangeCurser[8][1],sorted_rangeCurser[9][1]]
eight_Layer_tag=sorted_rangeCurser[8][0]

nineth_Layer_curser=[sorted_rangeCurser[9][1],sorted_rangeCurser[10][1]]
nineth_Layer_tag=sorted_rangeCurser[9][0]

tenth_Layer_curser=[sorted_rangeCurser[10][1],len(data)]
tenth_Layer_tag=sorted_rangeCurser[10][0]


print(first_Layer_curser)
print(second_Layer_curser)
print(nineth_Layer_curser)
print(tenth_Layer_tag)

for i in range(first_Layer_curser[0],first_Layer_curser[1]):
    data[i][0]=first_Layer_tag

for i in range(second_Layer_curser[0],second_Layer_curser[1]):
    data[i][0]=second_Layer_tag

for i in range(third_Layer_curser[0],third_Layer_curser[1]):
    data[i][0]=third_Layer_tag
    
for i in range(fourth_Layer_curser[0],fourth_Layer_curser[1]):
    data[i][0]=fourth_Layer_tag

for i in range(fifth_Layer_curser[0],fifth_Layer_curser[1]):
    data[i][0]=fifth_Layer_tag

for i in range(sixth_Layer_curser[0],sixth_Layer_curser[1]):
    data[i][0]=sixth_Layer_tag

for i in range(seventh_Layer_curser[0],seventh_Layer_curser[1]):
    data[i][0]=seventh_Layer_tag

for i in range(eight_Layer_curser[0],eight_Layer_curser[1]):
    data[i][0]=eight_Layer_tag

for i in range(nineth_Layer_curser[0],nineth_Layer_curser[1]):
    data[i][0]=nineth_Layer_tag

for i in range(tenth_Layer_curser[0],tenth_Layer_curser[1]):
    data[i][0]=tenth_Layer_tag
    
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#permanent clean of unwanted rows in dataset
tobedeletedList=[]
for row_num in range (1,len(data)):
    if (len(data[row_num])<3):#regarding temporary header size
        #print ("row num{}:{}".format(i,data[row_num]))
        tobedeletedList.append(row_num)
print (tobedeletedList)
for aggregated_row_num in tobedeletedList:
    #print (data[aggregated_row_num])
    del data[aggregated_row_num]
    tobedeletedList[:] = [x - 1 for x in tobedeletedList]
   #print (tobedeletedList)
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

<h4>construct  query</h4>
<p>EEOBY_Query=PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>\n\<br>
PREFIX OGI: <http://ogi.eu/#>\n\<br>
PREFIX qb: <http://purl.org/linked-data/cube#>\n\<br>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>\n\<br>
CONSTRUCT{\n\<br>
		?observation a qb:Observation;\n\<br>
				OGI:employee_number ?employee_number;\n\<br>
				OGI:municipality ?Municipality;\n\<br>
				OGI:time_period ?Time_period;\n\<br>
				OGI:economic_entities_in_operation_at_the_beginning_of_the_year_units ?Economic_entities_in_operation_at_the_beginning_of_the_year_units;\n\<br>
				OGI:counrty ?Country;\n\<br>
				OGI:county ?County;\n\<br>
				qb:dataSet OGI:EEOBY_ds;\n\<br>
				.\n\<br>
	}\n\<br>
WHERE {\n\<br>
BIND (uri(CONCAT('http://ogi.eu/#observations_',StrUUID())) AS ?observation).\n\<br>
}\n\<br>
