```python
#pulling csv data from file
import csv
filename='Source_datasets/Exported matrix table_d2.csv'
with open(filename, 'r') as f:
    data = list(csv.reader(f))
#print (data)
#print (f)
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#remove aggregated rows, as aggregation will take place later e.g. in pivotjs
keysList=['Republic','county']
tobedeletedList=[]
ageCurser={}#to tag age layers 
updated_lengh=len(data)
for row_num in range(0,updated_lengh):
    for key in keysList:
        #print (key)
        if key in data[row_num][1]:
            tobedeletedList.append(row_num)#for males row
            tobedeletedList.append(row_num+1)#for the females row
            if 'Republic' in key:
                ageCurser[data[row_num][0]]=row_num
                
import operator

sorted_ageCurser = sorted(ageCurser.items(), key=operator.itemgetter(1))

print (tobedeletedList)
print (ageCurser)
print('sorted cursor:{}'.format(sorted_ageCurser))


for aggregated_row_num in tobedeletedList:
    #print (data[aggregated_row_num])
    data[aggregated_row_num]=['']
    #tobedeletedList[:] = [x - 1 for x in tobedeletedList]
   #print (tobedeletedList)
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#remove first layer as it is aggregated (total by age)
totalbyAgeLayer_start_end=[sorted_ageCurser[0][1],sorted_ageCurser[1][1]]
print (totalbyAgeLayer_start_end)

del_curser=totalbyAgeLayer_start_end[1]-1 #-1 is important not to delete the start of the next layer
print (del_curser)

for i in range(totalbyAgeLayer_start_end[0],totalbyAgeLayer_start_end[1]):
    print ("delete row {}:{}".format(del_curser,data[del_curser]))
    data[del_curser]=['']
    del_curser=del_curser-1
    
for row_num in range(0,len(data)):
    print (data[row_num])

```

<p>sorted cursor:[('Total by age', 1), ('Population (0–15 years)', 145), ('Working-age population', 289), ('Pension age population', 433)]</p>

```python
#add missing Age tag layers values
firstAgeLayer_curser=[sorted_ageCurser[1][1],sorted_ageCurser[2][1]]
firstAgeLayer_tag=sorted_ageCurser[1][0]
secondAgeLayer_curser=[sorted_ageCurser[2][1],sorted_ageCurser[3][1]]
secondAgeLayer_tag=sorted_ageCurser[2][0]
thirdAgeLayer_curser=[sorted_ageCurser[3][1],len(data)]
thirdAgeLayer_tag=sorted_ageCurser[3][0]

print(firstAgeLayer_curser)
print(secondAgeLayer_curser)
print(thirdAgeLayer_curser)
print(firstAgeLayer_tag)

for i in range(firstAgeLayer_curser[0],firstAgeLayer_curser[1]):
    data[i][0]=firstAgeLayer_tag

for i in range(secondAgeLayer_curser[0],secondAgeLayer_curser[1]):
    data[i][0]=secondAgeLayer_tag

for i in range(thirdAgeLayer_curser[0],thirdAgeLayer_curser[1]):
    data[i][0]=thirdAgeLayer_tag
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#permanent clean of unwanted rows in dataset
tobedeletedList=[]
for row_num in range (1,len(data)):
    if (len(data[row_num])<3):#regarding temporary header size
        #print ("row num{}:{}".format(i,data[row_num]))
        tobedeletedList.append(row_num)
print (tobedeletedList)
for aggregated_row_num in tobedeletedList:
    #print (data[aggregated_row_num])
    del data[aggregated_row_num]
    tobedeletedList[:] = [x - 1 for x in tobedeletedList]
   #print (tobedeletedList)
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#add missing country values
for row_num in range(1,len(data)):
    data[row_num].append('Republic of Lithuania')
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#getting new county ranges (old dif*2+new)
def update_range(old_start,old_end,new_start):
    new_end_start=new_start+(old_end-old_start)*2
    return new_end_start

print (update_range(54,62,107))

```

```python
#add missing county values
chuncks=[0,122,244]

for chunck in range(0,3):
    for row_num in range(1+chuncks[chunck],11+chuncks[chunck]):
            data[row_num].append('Alytus county')
    for row_num in range(11+chuncks[chunck],27+chuncks[chunck]):
            data[row_num].append('Kaunas county')
    for row_num in range(27+chuncks[chunck],41+chuncks[chunck]):
            data[row_num].append('Klaipeda county')
    for row_num in range(41+chuncks[chunck],53+chuncks[chunck]):
            data[row_num].append('Marijampole county')
    for row_num in range(53+chuncks[chunck],65+chuncks[chunck]):
            data[row_num].append('Panevezys county')
    for row_num in range(65+chuncks[chunck],79+chuncks[chunck]):
            data[row_num].append('Siauliai county')
    for row_num in range(79+chuncks[chunck],87+chuncks[chunck]):
            data[row_num].append('Taurage county')
    for row_num in range(87+chuncks[chunck],95+chuncks[chunck]):
            data[row_num].append('Telsiai county')
    for row_num in range(95+chuncks[chunck],107+chuncks[chunck]):
            data[row_num].append('Utena county')
    for row_num in range(107+chuncks[chunck],123+chuncks[chunck]):
            data[row_num].append('Vilnius county')

        
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#Adding missing municapility value for the Females rows
for row_num in range(1,len(data)):
    if not data[row_num][1]:
        #print (data[row_num])
        data[row_num][1]=data[row_num-1][1]
        #print (data[row_num])
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#Adding collumn headers istead of the melformed header
headerList=['Age','Municipality','Sex','Time_period','Resident_population_at_the_beginning_of_the_year_persons','Country','County']
del data[0]
data.insert( 0, headerList)
for row_num in range(0,len(data)):
    print (data[row_num])
```

<h4>Construct query</h4>
<p>
RPBYmunicipality_Query=PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>\n\ <br>
PREFIX OGI: <http://ogi.eu/#>\n\<br>
PREFIX qb: <http://purl.org/linked-data/cube#>\n\<br>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>\n\<br>
CONSTRUCT{\n\<br>
		?observation a qb:Observation;\n\<br>
				OGI:age ?Age;\n\<br>
				OGI:municipality ?Municipality;\n\<br>
				OGI:sex ?Sex;\n\<br>
				OGI:time_period ?Time_period;\n\<br>
				OGI:resident_population_at_the_beginning_of_the_year_persons<br> ?Resident_population_at_the_beginning_of_the_year_persons;\n\<br>
				OGI:counrty ?Country;\n\<br>
				OGI:county ?County;\n\<br>
				qb:dataSet OGI:RPBYmunicipality_ds;\n\<br>
				.\n\<br>
	}\n\<br>
WHERE {\n\<br>
BIND (uri(CONCAT('http://ogi.eu/#observations_',StrUUID())) AS ?observation).\n\<br>
}\n\<br>
