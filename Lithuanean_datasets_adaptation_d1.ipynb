```python
#pulling csv data from file
#data sources :
#1.      Resident population at the beginning of the year (persons by municipality) - http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=7ed668c3-1fed-4990-8ff9-e1145867bb26
#2.      Resident population at the beginning of the year (by gender)- http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=e62988b7-de7a-4c72-83b3-9cad28f4f250
#3.      Resident population at the beginning of the year (by age groups)- http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=506b3efb-c5fa-4476-8926-780f95e4881f
#4.      Economic entities in operation at the beginning of the year (Total by employees) - http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=62026300-429b-41af-9712-3070f6ad8868
#5.      Foreign direct investment per capita at the end of the year (2014) - http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=b3226318-f641-4d66-847f-b9df5d05eb30
#6.      Investment in tangible fixed assets at current prices (2014) - http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=061b4c9b-03d7-4303-84b0-2a39bc730e00
#7.      Average earnings (monthly (2015))- http://osp.stat.gov.lt/en/statistiniu-rodikliu-analize?portletFormName=visualization&hash=331ea045-2544-49a2-a495-a99b6758413b
import csv
filename='Source_datasets/Exported matrix table_d1.csv'
with open(filename, 'r') as f:
    data = list(csv.reader(f))
#print (data)
#print (f)
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#Permanent remove aggregated rows, as aggregation will take place later e.g. in pivotjs
keysList=['Republic','county']
tobedeletedList=[]
updated_lengh=len(data)
for row_num in range(0,updated_lengh):
    for key in keysList:
        #print (key)
        if key in data[row_num][1]:
            tobedeletedList.append(row_num)

print (tobedeletedList)

for aggregated_row_num in tobedeletedList:
    #print (data[aggregated_row_num])
    del data[aggregated_row_num]
    tobedeletedList[:] = [x - 1 for x in tobedeletedList]
   #print (tobedeletedList)
    
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#add missing aggregation type values 
for row_num in range(1,len(data)):
    data[row_num][0]='Total by age'
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#add missing country values
for row_num in range(1,len(data)):
    data[row_num].append('Republic of Lithuania')
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#add missing county values
for row_num in range(1,6):
        data[row_num].append('Alytus county')
for row_num in range(6,14):
        data[row_num].append('Kaunas county')
for row_num in range(14,21):
        data[row_num].append('Klaipeda county')
for row_num in range(21,27):
        data[row_num].append('Marijampole county')
for row_num in range(27,33):
        data[row_num].append('Panevezys county')
for row_num in range(33,40):
        data[row_num].append('Siauliai county')
for row_num in range(40,44):
        data[row_num].append('Taurage county')
for row_num in range(44,48):
        data[row_num].append('Telsiai county')
for row_num in range(48,54):
        data[row_num].append('Utena county')
for row_num in range(54,62):
        data[row_num].append('Vilnius county')

        
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
#Adding collumn headers istead of the melformed header
headerList=['Age','Municipality','Sex','Time_period','Resident_population_at_the_beginning_of_the_year_persons','Country','County']
del data[0]
data.insert( 0, headerList)
for row_num in range(0,len(data)):
    print (data[row_num])
```

```python
import csv
#print (data)
with open("Ready_datasets/readyDataSet.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

<h4>Construct query</h4>
<p>
RPBYmunicipality_Query=PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>\n\ <br>
PREFIX OGI: <http://ogi.eu/#>\n\<br>
PREFIX qb: <http://purl.org/linked-data/cube#>\n\<br>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>\n\<br>
CONSTRUCT{\n\<br>
		?observation a qb:Observation;\n\<br>
				OGI:age ?Age;\n\<br>
				OGI:municipality ?Municipality;\n\<br>
				OGI:sex ?Sex;\n\<br>
				OGI:time_period ?Time_period;\n\<br>
				OGI:resident_population_at_the_beginning_of_the_year_persons<br> ?Resident_population_at_the_beginning_of_the_year_persons;\n\<br>
				OGI:counrty ?Country;\n\<br>
				OGI:county ?County;\n\<br>
				qb:dataSet OGI:RPBYmunicipality_ds;\n\<br>
				.\n\<br>
	}\n\<br>
WHERE {\n\<br>
BIND (uri(CONCAT('http://ogi.eu/#observations_',StrUUID())) AS ?observation).\n\<br>
}\n\<br>
